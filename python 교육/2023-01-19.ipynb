## 재귀함수 : 자기 자신을 호출하는 함수

def recur():
    print("뿅")
    recur()


```python
def my_magic_func(n):
    return n * 10

my_list = [1, 2, 3, 4, 5]
a = []
for i in range(1, len(my_list)+1):
    
    a.append(i*10)
print(a)

result = list(map(my_magic_func, my_list))
print(result)

# zip
name_ = ["신동민", "서재현", "박영서", "이태성","정예원", "이은석"]
age_ = [12, 18, 22, 24, 25, 19]

for each in zip(name_, age_):
    print(each)

for name, age in zip(name_, age_):
    print(name, age)

# lambda

def my_magic_func(n):
    return n * 10

result = list(map(my_magic_func, [1, 2, 3]))
print(result)

result_ = list(map((lambda x: x * 10), [1, 2, 3]))
print(result_)


print((lambda x: x * x)(4))

myfunc_ = lambda n: n * 2
print(myfunc_(2))

def pow(x):
    return x * x


# 재귀함수

def recur():
    print("뿅")
    recur()

recur()
```
