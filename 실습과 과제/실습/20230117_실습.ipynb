```python
# 실습2

steak = 50000
print(f"스테이크{steak:>10}")
print(f"스테이크{steak:<10}")
print(f"스테이크{steak:^10}")

# steak뒤에다가
#  :>len 를 쓰면 오른쪽 정렬, len만큼 확보 
#  :<len 를 쓰면 왼 정렬, len만큼 확보
#  :^len 를 쓰면 가운데 정렬, len만큼 확보
```
